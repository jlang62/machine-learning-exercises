```python
non_null_counts = df.count()
columns_of_small_nulls = non_null_counts[non_null_counts >= 700000].index
df1= df[columns_of_small_nulls]
print(df.columns.tolist())
df1=df1.dropna()
```

```python
df1[['LATITUDE','LONGITUDE']].value_counts()[0:24]
```

```python
value_counts = df1[['LATITUDE', 'LONGITUDE']].value_counts()
valid_combinations = value_counts[1:23].index
print(valid_combinations)
```

```python
dfi = df1[df1[['LATITUDE', 'LONGITUDE']].apply(tuple, axis=1).isin(valid_combinations)]
ltm,lgm=dfi['LATITUDE'].mean(),dfi['LONGITUDE'].mean()

eq_map = folium.Map(location=[ltm,lgm],tiles='Stamen Terrain',zoom_start=10.0,min_zoom=2.0)
eq_map.add_child(plugins.HeatMap(dfi[['LATITUDE','LONGITUDE']]))
eq_map
```
