```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./data/F1Drivers_Dataset.csv')
```

```python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Plot 1: Bar Plot for Nationality Counts
nationality_counts = df['Nationality'].value_counts()
nationality_counts.plot(kind='bar', ax=axes[0, 0])
axes[0, 0].set_xlabel('Nationality')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_title('Nationality Counts')

# Plot 2: Histogram for Race Starts
axes[0, 1].hist(df['Race_Starts'], bins=20, edgecolor='k')
axes[0, 1].set_xlabel('Race Starts')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].set_title('Race Starts Histogram')

# Plot 3: Box Plot for Race Wins
axes[1, 0].boxplot(df['Race_Wins'])
axes[1, 0].set_xlabel('Race Wins')
axes[1, 0].set_title('Race Wins Box Plot')

# Plot 4: Scatter Plot for Win Rate vs. Podium Rate
axes[1, 1].scatter(df['Win_Rate'], df['Podium_Rate'])
axes[1, 1].set_xlabel('Win Rate')
axes[1, 1].set_ylabel('Podium Rate')
axes[1, 1].set_title('Win Rate vs. Podium Rate')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
